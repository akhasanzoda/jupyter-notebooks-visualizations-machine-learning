```python
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
file_path = 'labor market.csv'
labor_market_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
labor_market_data.head()
```

```python
# Remove the "Unnamed: 32" column
labor_market_data_cleaned = labor_market_data.drop(columns=['Unnamed: 32'])

# Display a sample of the cleaned data to ensure it looks correct
labor_market_data_cleaned.head()
```

```python
labor_market_data_cleaned = labor_market_data_cleaned.dropna(how='all', subset=labor_market_data_cleaned.columns[1:])

# Display a sample of the data to ensure the row has been removed correctly
labor_market_data_cleaned.head()
```

```python
# Generate summary statistics for the dataset
summary_statistics = labor_market_data_cleaned.describe()

# Transpose for a better view
summary_statistics.T
```

```python
# Filter rows for economically active population and employed population
economically_active = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'] == 'Economically active population ']
employed_population = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'] == 'Number of Employed Population ']

# Prepare data for plotting
years_quarters = economically_active.columns[1:]  # Exclude 'Descriptor' column
economically_active_values = economically_active.iloc[0, 1:].values
employed_population_values = employed_population.iloc[0, 1:].values

# Plotting
plt.figure(figsize=(14, 7), dpi=200)

# Economically active population
plt.plot(years_quarters, economically_active_values, label='Economically Active Population', marker='o')

# Employed population
plt.plot(years_quarters, employed_population_values, label='Employed Population', marker='x')

# Enhancing the plot
plt.title('Labor Market Trends in Uzbekistan')
plt.xlabel('Year / Quarter')
plt.ylabel('Population (in thousands)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

```python
sectors_to_visualize = [
    'Agriculture, forestry and fisheries',
    'Industry',
    'Construction',
    'Trade',
    'Transportation and Storage',
    'Education',
    'Health and social services',
    'Other activities'
]

# Filtering the dataset for the specified sectors
visualize_sector_data = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(sectors_to_visualize)]

# Plotting
plt.figure(figsize=(14, 10), dpi=200)

for sector in sectors_to_visualize:
    sector_values = visualize_sector_data[visualize_sector_data['Descriptor'] == sector].iloc[0, 1:].values
    plt.plot(years_quarters, sector_values, label=sector, marker='o')

# Enhancing the plot
plt.title('Employment Trends for Specified Sectors in Uzbekistan')
plt.xlabel('Year / Quarter')
plt.ylabel('Employed Population (in thousands)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```

```python
sectors_excluded_visualize_corrected = [
    'Public sector',
    'Non-Public sector',
    'Unemployed',
    'Registered in labor agencies*'
]
# Changing the descriptor of the last row to "Unemployed"
labor_market_data_cleaned.at[labor_market_data_cleaned.index[-1], 'Descriptor'] = 'Unemployed'

if 'Unemployed' not in sectors_excluded_visualize_corrected:
    sectors_excluded_visualize_corrected.append('Unemployed')
excluded_sector_data_updated = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(sectors_excluded_visualize_corrected)]

plt.figure(figsize=(14, 10), dpi=200)

for sector in sectors_excluded_visualize_corrected:
    if sector in excluded_sector_data_updated['Descriptor'].values:
        sector_values = excluded_sector_data_updated[excluded_sector_data_updated['Descriptor'] == sector].iloc[0, 1:].values
        plt.plot(years_quarters, sector_values, label=sector, marker='o')

# Enhancing the plot
plt.title('Employment Trends in Uzbekistan')
plt.xlabel('Year / Quarter')
plt.ylabel('Number (in thousands)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```

```python
# Isolating the data for the "Unemployed" category
unemployed_data = excluded_sector_data_updated[excluded_sector_data_updated['Descriptor'] == 'Unemployed'].iloc[0, 1:].values

# Plotting the "Unemployed" category
plt.figure(figsize=(14, 7), dpi=200)

plt.plot(years_quarters, unemployed_data, label='Unemployed', marker='o', color='red')

# Enhancing the plot
plt.title('Unemployment Trends in Uzbekistan')
plt.xlabel('Year / Quarter')
plt.ylabel('Number of Unemployed (in thousands)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Extracting the final year (or most recent data point) column name
final_year_column = labor_market_data_cleaned.columns[-1]

# Separating the dataset into actual sectors and broader categories
actual_sectors = [
    'Agriculture, forestry and fisheries',
    'Industry',
    'Construction',
    'Trade',
    'Transportation and Storage',
    'Education',
    'Health and social services',
    'Other activities'
]

broader_categories = [
    'Public sector',
    'Non-Public sector',
    'Unemployed'
]

# Extracting final year values for actual sectors
final_year_values_sectors = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(actual_sectors)][['Descriptor', final_year_column]]

# Extracting final year values for broader categories
final_year_values_categories = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(broader_categories)][['Descriptor', final_year_column]]

# Visualizing actual sectors
plt.figure(figsize=(12, 8), dpi=200)
plt.barh(final_year_values_sectors['Descriptor'], final_year_values_sectors[final_year_column], color='skyblue')
plt.title('Employment in Actual Sectors for the Final Year')
plt.xlabel('Number of Employed (in thousands)')
plt.ylabel('Sectors')
plt.tight_layout()
plt.grid(True)
plt.show()

# Visualizing broader categories
plt.figure(figsize=(12, 6), dpi=200)
plt.barh(final_year_values_categories['Descriptor'], final_year_values_categories[final_year_column], color='salmon')
plt.title('Employment in Broader Categories for the Final Year')
plt.xlabel('Number of Employed/Unemployed (in thousands)')
plt.ylabel('Categories')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
first_year_column = labor_market_data_cleaned.columns[1]

sector_growth_rates = []
for index, row in labor_market_data_cleaned.iterrows():
    first_year_value = row[first_year_column]
    final_year_value = row[final_year_column]
    if first_year_value > 0:  # To avoid division by zero
        growth_rate = ((final_year_value - first_year_value) / first_year_value) * 100
    else:
        growth_rate = None  # Handling sectors with zero or no initial data
    sector_growth_rates.append((row['Descriptor'], first_year_value, final_year_value, growth_rate))

sector_growth_df = pd.DataFrame(sector_growth_rates, columns=['Sector', 'First Year Value', 'Final Year Value', 'Growth Rate'])
actual_sector_growth_df = sector_growth_df[sector_growth_df['Sector'].isin(actual_sectors)]

# Recalculating the fastest growing and slowest-growing sectors
fastest_growing_actual_sector = actual_sector_growth_df.loc[actual_sector_growth_df['Growth Rate'].idxmax()]
slowest_growing_actual_sector = actual_sector_growth_df.loc[actual_sector_growth_df['Growth Rate'].idxmin()]

# Visualizing the fastest and slowest growing sectors
plt.figure(figsize=(10, 6), dpi=200)
bar_colors = ['green' if sector == fastest_growing_actual_sector['Sector'] else 'red' if sector == slowest_growing_actual_sector['Sector'] else 'grey' for sector in actual_sector_growth_df['Sector']]
plt.barh(actual_sector_growth_df['Sector'], actual_sector_growth_df['Growth Rate'], color=bar_colors)
plt.xlabel('Growth Rate (%)')
plt.ylabel('Sector')
plt.title('Growth Rate of Actual Economic Sectors in Uzbekistan')
plt.axvline(x=0, color='black', linewidth=0.8)  # Zero line for reference

# Highlighting the fastest and slowest growing sectors
for index, value in enumerate(actual_sector_growth_df['Growth Rate']):
    if actual_sector_growth_df.iloc[index]['Sector'] in [fastest_growing_actual_sector['Sector'], slowest_growing_actual_sector['Sector']]:
        plt.text(value, index, f"{value:.1f}%", va='center', font='Arial', fontsize=10, color='black')
plt.tight_layout()
plt.show()

fastest_growing_actual_sector, slowest_growing_actual_sector
```

```python
final_year_values_actual_sectors_corrected = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(actual_sectors)][['Descriptor', final_year_column]]

# Pie chart preparation with the corrected data
labels_corrected = final_year_values_actual_sectors_corrected['Descriptor']
sizes_corrected = final_year_values_actual_sectors_corrected[final_year_column]
explode_corrected = [0.1 if size == sizes_corrected.max() else 0 for size in sizes_corrected]  # Explode the largest sector

# Pie chart visualization with the corrected data
plt.figure(figsize=(10, 8), dpi=200)
plt.pie(sizes_corrected, explode=explode_corrected, labels=labels_corrected, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='black'))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Employment Across Sectors in the Final Year')
plt.show()

# Identifying the largest sector for reference with the corrected data
largest_sector_corrected = final_year_values_actual_sectors_corrected.loc[final_year_values_actual_sectors_corrected[final_year_column].idxmax()]
largest_sector_corrected
```

```python
# Using a specific color for the largest sector and a neutral color for the others
colors = ['lightblue' if size != sizes_corrected.max() else 'orange' for size in sizes_corrected]

# Pie chart visualization with specified colors
plt.figure(figsize=(10, 8), dpi=200)
plt.pie(sizes_corrected, explode=explode_corrected, labels=labels_corrected, autopct='%1.1f%%', colors=colors, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Employment Across Sectors in the Final Year with Specified Colors')
plt.show()
```

```python
# Extracting employment figures for 2019, 2020, and 2021 for actual sectors
covid_impact_years = ['2019', '2020', '2021']
covid_impact_data = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'].isin(actual_sectors)][['Descriptor'] + covid_impact_years]

# Calculating year-over-year changes for 2019-2020 and 2020-2021
covid_impact_data['2019-2020 Change'] = (covid_impact_data['2020'] - covid_impact_data['2019']) / covid_impact_data['2019'] * 100
covid_impact_data['2020-2021 Change'] = (covid_impact_data['2021'] - covid_impact_data['2020']) / covid_impact_data['2020'] * 100

# Visualizing the impact of COVID-19 on employment in sectors
plt.figure(figsize=(14, 10), dpi=200)

# 2019-2020 Changes
plt.subplot(1, 2, 1)
plt.barh(covid_impact_data['Descriptor'], covid_impact_data['2019-2020 Change'], color='skyblue')
plt.title('2019-2020 Employment Change (%)')
plt.xlabel('Percentage Change')
plt.axvline(x=0, color='black', linewidth=0.8)  # Zero line for reference
plt.grid(True)

# 2020-2021 Changes
plt.subplot(1, 2, 2)
plt.barh(covid_impact_data['Descriptor'], covid_impact_data['2020-2021 Change'], color='lightgreen')
plt.title('2020-2021 Employment Change (%)')
plt.xlabel('Percentage Change')
plt.axvline(x=0, color='black', linewidth=0.8)  # Zero line for reference
plt.grid(True)
plt.tight_layout()
plt.show()

covid_impact_data[['Descriptor', '2019-2020 Change', '2020-2021 Change']]
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

public_sector_data = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'] == 'Public sector'].iloc[0, 8:-1]
non_public_sector_data = labor_market_data_cleaned[labor_market_data_cleaned['Descriptor'] == 'Non-Public sector'].iloc[0, 8:-1]

# Converting to numeric and handling potential non-numeric entries
public_sector_data = pd.to_numeric(public_sector_data, errors='coerce').dropna()
non_public_sector_data = pd.to_numeric(non_public_sector_data, errors='coerce').dropna()

# Public Sector Forecast
arima_public = ARIMA(public_sector_data, order=(1, 1, 1))  # (p,d,q) parameters are chosen based on typical ARIMA modeling practices
arima_public_fit = arima_public.fit()
public_forecast = arima_public_fit.forecast(steps=4)  # Forecasting the next 4 quarters

# Non-Public Sector Forecast
arima_non_public = ARIMA(non_public_sector_data, order=(1, 1, 1))  # Adjusting parameters may improve forecast accuracy
arima_non_public_fit = arima_non_public.fit()
non_public_forecast = arima_non_public_fit.forecast(steps=4)  # Forecasting the next 4 quarters

# Visualizing the Forecast
plt.figure(figsize=(14, 7), dpi=200)

# Public Sector
plt.subplot(1, 2, 1)
plt.plot(public_sector_data.index, public_sector_data, label='Historical Data', marker='o')
plt.plot(np.arange(len(public_sector_data), len(public_sector_data) + 4), public_forecast, label='Forecast', marker='x', linestyle='--')
plt.title('Public Sector Employment Forecast')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Non-Public Sector
plt.subplot(1, 2, 2)
plt.plot(non_public_sector_data.index, non_public_sector_data, label='Historical Data', marker='o')
plt.plot(np.arange(len(non_public_sector_data), len(non_public_sector_data) + 4), non_public_forecast, label='Forecast', marker='x', linestyle='--')
plt.title('Non-Public Sector Employment Forecast')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
